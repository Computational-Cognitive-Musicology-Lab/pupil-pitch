```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('main_final_2.csv')
```

```python
data.head()
```

```python
df = data.loc[:, [False, True, False, True, True,True,True,True,True,True,True,True,True,True, True,True,True,True,True, False]]
df.head()
#df.groupby('Participant')
```

```python
vals = df.groupby('Participant').mean()
```

```python
vals.head()
```

```python
#throw out 'reject' row
vals = vals.iloc[:-1,:]
```

```python
#all participants
vals.transpose().plot()
```

```python
#plot of absolute values means + sd for each participant for ALL Hz presentations
vals.transpose().mean(axis=1).plot(rot=90)
```

```python
rel_vals = vals.transpose() - (vals.transpose().mean(axis=0))
```

```python
#individuals (means above)
rel_vals.plot()
```

```python
ci = 1.96 * vals.transpose().std(axis=1)/np.sqrt(len(rel_vals))
```

```python
participant_means = list(vals.transpose().mean(axis=1))

fig, ax = plt.subplots()
ax.plot(vals.transpose().mean(axis=1))
ax.fill_between(vals.transpose().index, (participant_means-ci), (participant_means+ci), color='b', alpha=.1)
plt.xticks(rotation=90)
```
